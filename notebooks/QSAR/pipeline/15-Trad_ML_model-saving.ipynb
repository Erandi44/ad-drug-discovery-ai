
# **Train and Save Final Traditional ML Model**s

This notebook retrains and serializes the final traditional machine learning (ML) models for each target using hyperparameters selected via Bayesian optimization within a nested cross-validation framework.

For each target:

- The reduced REU feature dataset is loaded and split into predefined training and external test sets.

- The best-performing hyperparameters (based on outer-loop RMSE from nested cross-validation) are applied to reconstruct the corresponding sklearn.pipeline.Pipeline.

- Each model is retrained on the full training dataset to maximize data utilization after unbiased model selection.

- The trained pipelines are saved as .joblib files for reproducible inference and downstream deployment.

- The exact feature column order is stored separately to ensure consistent feature alignment when scoring new compounds.

- This notebook serves as the artifact-generation stage of the QSAR workflow, producing reusable, deployment-ready traditional ML models without using external test data during training.

```python
# make_model() — this function returns a sklearn.pipeline.Pipeline

def make_model(model_name: str):
    model_name = model_name.strip()

    if model_name == "RandomForest":
        model = RandomForestRegressor(random_state=42)
        pipe = Pipeline([("model", model)])
        return pipe

    if model_name == "GradientBoosting":
        model = GradientBoostingRegressor(random_state=42)
        pipe = Pipeline([("model", model)])
        return pipe

    if model_name == "XGBoost":
        if not HAS_XGB:
            raise RuntimeError("XGBoost not installed/importable in this environment.")
        #  initialize the model with safe base settings. The optimized hyperparameters will replace them
        model = XGBRegressor(
            random_state=42,
            objective="reg:squarederror",
            n_jobs=-1
        )
        pipe = Pipeline([("model", model)])
        return pipe

    if model_name == "SVR":
        model = SVR()
        pipe = Pipeline([("scaler", StandardScaler()), ("model", model)])
        return pipe

    if model_name == "KNeighborsRegressor":
        model = KNeighborsRegressor()
        pipe = Pipeline([("scaler", StandardScaler()), ("model", model)])  # KNN sensitive to scale
        return pipe

    if model_name == "MLPRegressor":
        model = MLPRegressor(
            random_state=42,
            max_iter=2000,
            early_stopping=True
        )
        pipe = Pipeline([("scaler", StandardScaler()), ("model", model)])
        return pipe

    raise ValueError(f"Unknown model: {model_name}")
```

```python
import os
import joblib
import pandas as pd

# NEW folder for predictions (won't overwrite old)
pred_dir_v2 = "/content/drive/MyDrive/Colab Notebooks/ESOFT-MSC/thesis-project/trad-ml-predictions/preds2"
os.makedirs(pred_dir_v2, exist_ok=True)

# folder for saved models
os.makedirs(model_dir, exist_ok=True)

def train_save_models_and_preds_newfolder(
    targets, reu_path, pred_dir_v2, model_dir, meta_cols, best_params
):
    for target in targets:
        print(f"\n===== Target: {target} =====")

        # Load train/test
        df_train = pd.read_csv(f"{reu_path}/{target}_train_reu.csv").dropna(how="any")
        df_test  = pd.read_csv(f"{reu_path}/{target}_test_reu.csv").dropna(how="any")

        X_train = df_train.drop(columns=meta_cols)
        y_train = df_train["pIC50"].values
        X_test  = df_test.drop(columns=meta_cols)
        y_test  = df_test["pIC50"].values

        # Save feature columns (needed later for new molecule prediction)
        feat_fp = os.path.join(model_dir, f"{target}__feature_columns.joblib")
        joblib.dump(list(X_train.columns), feat_fp)

        # Prepare outputs (these are NEW files in pred_dir_v2)
        test_out = df_test[["molecule_chembl_id", "canonical_smiles", "pIC50"]].copy()
        train_out = df_train[["molecule_chembl_id", "canonical_smiles", "pIC50"]].copy()

        for model_name, params in best_params[target].items():
            try:
                pipe = make_model(model_name)
                pipe.set_params(**params)
                pipe.fit(X_train, y_train)

                # Save model
                model_fp = os.path.join(model_dir, f"{target}__{model_name}.joblib")
                joblib.dump(pipe, model_fp)

                # Predict
                test_out[f"pred_{model_name}"]  = pipe.predict(X_test)
                train_out[f"pred_{model_name}"] = pipe.predict(X_train)

                print(f"[OK] {target} - {model_name}")

            except Exception as e:
                print(f"[FAIL] {target} - {model_name}: {e}")

        # Save preds to NEW folder (no overwrite of old folder)
        test_csv  = os.path.join(pred_dir_v2, f"{target}_test_preds_tradML.csv")
        train_csv = os.path.join(pred_dir_v2, f"{target}_train_preds_tradML.csv")
        test_out.to_csv(test_csv, index=False)
        train_out.to_csv(train_csv, index=False)

        print("Saved preds:", test_csv)
        print("Saved preds:", train_csv)
        print("Saved feature cols:", feat_fp)

# Run
train_save_models_and_preds_newfolder(
    targets=targets,
    reu_path=reu_path,
    pred_dir_v2=pred_dir_v2,
    model_dir=model_dir,
    meta_cols=meta_cols,
    best_params=best_params
)
```
