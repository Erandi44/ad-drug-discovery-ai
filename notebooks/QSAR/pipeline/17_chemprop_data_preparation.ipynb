# **Preparing Input Data for Chemprop Modeling**

This notebook extracts the SMILES strings (canonical_smiles) and corresponding pIC50 activity values from the cleaned training and test datasets. These are saved as CSV files with the required Chemprop format (smiles, pIC50) for model training and evaluation.

In addition, the notebook extracts the RFECV-selected descriptor features from the datasets by removing metadata columns. These reduced feature matrices are saved separately so they can be used as external molecular features during Chemprop training.

The resulting files provide both:

- Standard Chemprop input datasets (SMILES + target values)

- Additional descriptor feature matrices for use with Chemprop’s feature-based modeling options.

```python
# === Paths to datasets ===

reu_path = "/content/drive/MyDrive/Colab Notebooks/ESOFT-MSC/thesis-project/data/descriptors/reduced_features_reu_2"
meta_cols = ['molecule_chembl_id', 'canonical_smiles', 'activity_class', 'pIC50']
```

```python
targets = ["5-HT6","ache", "bace1", "buche","esr1","gsk-3beta", "mao-b"]

target=targets[6]
```

```python
import pandas as pd

df_train = pd.read_csv(f"{reu_path}/{target}_train_reu.csv").dropna(how="any")
df_test = pd.read_csv(f"{reu_path}/{target}_test_reu.csv").dropna(how="any")
```

```python
df_train.head()
```

```python


# Combine SMILES and pIC50 for training
train_df = pd.DataFrame({
    "smiles": df_train["canonical_smiles"],
    "pIC50": df_train["pIC50"]
})

# Combine SMILES and pIC50 for testing
test_df = pd.DataFrame({
    "smiles": df_test["canonical_smiles"],
    "pIC50": df_test["pIC50"]
})

# Save to CSV in Chemprop-compatible format
train_df.to_csv(f"/content/drive/MyDrive/Colab Notebooks/ESOFT-MSC/thesis-project/chemprop/input_data/{target}_train.csv", index=False)
test_df.to_csv(f"/content/drive/MyDrive/Colab Notebooks/ESOFT-MSC/thesis-project/chemprop/input_data/{target}_test.csv", index=False)

print("Saved 'train.csv' and 'test.csv' for Chemprop.")
```

```python
# Extract RFECV features from train and test sets
X_train_reu = df_train.drop(columns=meta_cols)
X_test_reu = df_test.drop(columns=meta_cols)

# Save to CSV
X_train_reu.to_csv(f"/content/drive/MyDrive/Colab Notebooks/ESOFT-MSC/thesis-project/chemprop/input_data/reu_features/{target}_train_reu.csv", index=False)
X_test_reu.to_csv(f"/content/drive/MyDrive/Colab Notebooks/ESOFT-MSC/thesis-project/chemprop/input_data/reu_features/{target}_test_reu.csv", index=False)
```
