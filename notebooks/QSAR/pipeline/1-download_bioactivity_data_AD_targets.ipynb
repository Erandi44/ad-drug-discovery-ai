 **Search for Target in ChEMBL**

Go to: https://www.ebi.ac.uk/chembl/

Search target protein name, e.g., "BACE1" or "Acetylcholinesterase".

Click on the most relevant result. When multiple ChEMBL entries exist for the same target,choose the most appropriate one.

1. Target Type: Prefer “Single Protein”
2. Organism: Choose Homo sapiens
3. Target Confidence Score (ChEMBL ≥27 only). If shown, prefer entries with a high confidence score (e.g., 9 or 10). This means: 9 = Direct single protein assignment, 8 or below = Less confident assignment (e.g., via homology)
4. Data Availability: Pick Entry with Most Bioactivity Data
Look under "Bioactivities" tab in the Target Report Card. Sort by “Assay type” or “Activity type” to see how many IC₅₀ records are available.
5. Assay Relevance (Optional, for advanced filtering)
Check if the assays are binding assays or functional assays.Binding (e.g., IC50) is usually better for QSAR modeling.Filter out low-quality or non-human assays if possible.

**Target search for Alzheimer - Beta-secretase 1, CHEMBL4822**

```python
target = new_client.target
#  Search for the target using ChEMBL ID
target_query = target.search('CHEMBL4822')
target1 = pd.DataFrame.from_dict(target_query)
target1
```

```python
selected_target = target1.target_chembl_id[0]
selected_target
```

```python
# Get IC50 activity
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
df1 = pd.DataFrame.from_dict(res)
```

```python
# Save raw data
df1.to_csv('/content/drive/MyDrive/Colab Notebooks/ESOFT-MSC/thesis-project/data/raw_bace1_ic50.csv', index=False)
```
