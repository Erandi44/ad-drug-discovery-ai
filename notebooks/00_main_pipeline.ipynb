# Alzheimer’s Disease Drug Discovery Workflow

This notebook provides an overview of the computational pipeline used in this project for Alzheimer’s disease drug discovery.

The workflow combines **QSAR modeling** and **Generative AI–based molecule design**.  
Instead of executing all steps from a single notebook, this notebook serves as a **navigation guide** to the individual notebooks that implement each stage of the pipeline.

---

# Overall Pipeline

Bioactivity Data Collection  
→ Data Cleaning and Preprocessing  
→ Exploratory Data Analysis (EDA)  
→ Descriptor Calculation and Feature Selection  
→ QSAR Model Development  
→ Ensemble Model Construction  
→ Generative AI Molecule Generation  
→ Medicinal Chemistry Filtering  
→ ADMET and Toxicity Screening
